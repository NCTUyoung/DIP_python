```python
%matplotlib inline
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
from scipy.linalg import solve
import seaborn as sns
sns.set_style("white")
im = Image.open('C:/Users/user/Documents/DIP/DIP3E_Original_Images_CH04/Fig0429(a)(blown_ic).tif')
im = np.asarray(im.convert('L'))
plt.imshow(im)
plt.gray()
plt.show()
ax = sns.distplot(im.flatten(), color="y")
```

```python
#1 填補原圖，成(2M,2N)
im_p = np.zeros((im.shape[0]*2,im.shape[1]*2))
im_p[:im.shape[0],:im.shape[1]] = im.copy()
plt.imshow(im_p)

plt.show()
# 平移頻率至中央
for i in range(im_p.shape[0]):
    for j in range(im_p.shape[1]):
        im_p[i][j] = im_p[i][j]*np.power(-1,i+j)
#計算DFT
fft_f = np.fft.fft2(im_p)
plt.imshow(1+np.log10(np.abs(fft_f)))
plt.show()
```

```python
# 產生實數濾波器H(u,v)

# 先給定空間濾波器h(x,y)
h = (1/16)*np.array([[1,2,1],[2,4,2],[1,2,1]])
print(h)
#填補成h_p(x,y) 大小(P,Q)
h_p = np.zeros((im.shape[0]*2,im.shape[1]*2))
h_p[:h.shape[0],:h.shape[1]] = h.copy()
# 頻域置中
for i in range(h_p.shape[0]):
    for j in range(h_p.shape[1]):
        h_p[i][j] = h_p[i][j]*np.power(-1,i+j)
#計算DFT
fft_h = np.fft.fft2(h_p)
plt.imshow(np.abs(fft_h))
plt.show()
```

```python
G = np.multiply(fft_f,fft_h)
g = np.real(np.fft.ifft2(G))
for i in range(g.shape[0]):
    for j in range(g.shape[1]):
        g[i][j] = g[i][j]*np.power(-1,i+j)
plt.imshow(g)
plt.show()
```

```python
crop = g[:im.shape[0],:im.shape[1]]
plt.imshow(crop)
plt.show()
```
