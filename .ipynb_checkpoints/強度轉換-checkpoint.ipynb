```python
%matplotlib inline
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
from scipy.linalg import solve
```

```python
def log_fun(c,r):
    return c*np.log2(1+r)
vfunc = np.vectorize(log_fun,excluded='c')
im = Image.open('fa.jpeg')
im = np.asarray(im.convert('L'))
new_pic = vfunc(1,im)
print(new_pic.shape)
plt.imshow(new_pic)
plt.gray()
plt.show()
```

```python
def gamma_fun(c,gamma,r):
    return c*np.power(r,gamma)
gfunc = np.vectorize(gamma_fun,excluded=['c,gamma'])
gamma = [0.1,0.3,0.7,1,1.5,2,3]
for ga in gamma:
    new_pic2 = gfunc(1,ga,im)
    plt.imshow(new_pic2)
    plt.gray()
    plt.show()


```
