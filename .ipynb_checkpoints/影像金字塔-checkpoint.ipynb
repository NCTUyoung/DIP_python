<H1>子頻帶編碼</H1>

```python
h1 = [-0.23037781,0.71484657,-0.63088076,0.02798376,0.18703484,-0.3084138,0.03288301,0.01059740]
h0 = [-0.01059740,0.03288301,0.3084138,-0.18703484,-0.02798376,0.63088076,0.71484657,0.23037781]
```

```python
a0 = np.zeros(im.shape)
a1 = np.zeros(im.shape)
b0 = np.zeros((im.shape[0]/2,im.shape[1]))
b1 = np.zeros((im.shape[0]/2,im.shape[1]))
b2 = np.zeros((im.shape[0]/2,im.shape[1]))
b3 = np.zeros((im.shape[0]/2,im.shape[1]))
for i in range(im.shape[1]):
    a0[:,i] = np.convolve(h0,im[:,i],mode='same')
    a1[:,i] = np.convolve(h1,im[:,i],mode='same')
#downsample on row
a0 = a0[::2,:]
a1 = a1[::2,:]
for i in range(im.shape[0]/2):
    b0[i,:] = np.convolve(h0,a0[i,:],mode='same')
    b1[i,:] = np.convolve(h1,a0[i,:],mode='same')
    b2[i,:] = np.convolve(h0,a1[i,:],mode='same')
    b3[i,:] = np.convolve(h1,a1[i,:],mode='same')
a = b0[:,::2]
dv = b1[:,::2]
dh = b2[:,::2]
dd = b3[:,::2]
    
```

```python
fig, axes = plt.subplots(2, 2)
axes[0,0].imshow(a)
axes[0,1].imshow(dv)
axes[1,0].imshow(dh)
axes[1,1].imshow(dd)
fig.show()
```
