```python
%matplotlib inline
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
from scipy.linalg import solve
import seaborn as sns


sns.set_style("white")
im = Image.open('fa.jpeg')
im = np.asarray(im.convert('L'))
plt.imshow(im)
plt.gray()
plt.show()
ax = sns.distplot(im.flatten(), color="y")

im2 = Image.open('C:/Users/user/Documents/DIP/DIP3E_Original_Images_CH04'+'/Fig0424(a)(rectangle).tif')
im2 = np.asarray(im2.convert('L'))
im3 = Image.open('C:/Users/user/Documents/DIP/DIP3E_Original_Images_CH04'+'/Fig0424(a)(rectangle_r).tif')
im3 = np.asarray(im3.convert('L'))
plt.imshow(im2)

plt.show()
```

```python
new_pic = np.zeros(im2.shape)
new_pic =  np.fft.fft2(im2)
x,y = np.mgrid[0:im2.shape[0],0:im2.shape[1]]
plt.imshow(np.abs(new_pic))
plt.show()
```

```python
# 原圖經過-1^x+y
shift_im2 = np.zeros(im2.shape)
for i in range(im2.shape[0]):
    for j in range(im2.shape[1]):
        shift_im2[i][j] = np.power(-1,i+j)*im2[i][j]
new_pic2 = np.zeros(shift_im2.shape)
new_pic2 =  np.fft.fft2(shift_im2)
```

```python
im2[512:612,512:612]
shift_im2[512:612,512:612]
plt.imshow(np.abs(new_pic2))
plt.show()
```

```python
# 執行log 轉換
new_pic_log = np.zeros(im2.shape)

for i in range(im2.shape[0]):
    for j in range(im2.shape[1]):
        new_pic_log[i][j] = 1+np.log10(new_pic2[i][j])
plt.imshow(new_pic_log)
plt.show()       
```
