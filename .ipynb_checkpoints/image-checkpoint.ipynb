```python
%matplotlib inline
import numpy as np
from scipy import misc
f = misc.face()


import matplotlib.pyplot as plt
plt.imshow(f)
plt.show()
f.shape
```

```python
noise = np.random.normal(0,10,(768, 1024, 3))
demo = f + noise
DEMO= demo.astype(np.uint8)
plt.imshow(DEMO)
plt.show()
for i in range(5):
    noise = np.random.normal(0,128,(768, 1024, 3))
    a = f + noise
    demo = demo + a
    DEMO= (demo/(i+2)).astype(np.uint8)
    
plt.imshow(DEMO)
plt.show()
```
