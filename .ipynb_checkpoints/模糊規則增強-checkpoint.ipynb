```python
im2 = Image.open(path+'/Fig0354(a)(einstein_orig).tif')
im2 = np.asarray(im2.convert('L'))
plt.imshow(im2)
plt.show()
```

```python
# 
mu_dark = np.zeros(256)
mu_gray = np.zeros(256)
mu_bright = np.zeros(256)
mu_dark[:64] = 1
for i in range(64,128):
    mu_dark[i] = 1-(i-63)*(1/64)
for i in range(64,128):
    mu_gray[i] = (i-63)*(1/64)
for i in range(128,192):
    mu_gray[i] = 1 - (i-127)*(1/64)
for i in range(128,192):
    mu_bright[i] = (i-127)*(1/64)
mu_bright[192:256] = 1
plt.plot(mu_dark,'b')
plt.plot(mu_bright,'g')
plt.plot(mu_gray,'r')
vd = 0
vg = 127
vb = 255
```

```python
def defuzzification(z):
    numerator = mu_dark[z]*vd+mu_gray[z]*vg+mu_bright[z]*vb
    denominator = mu_dark[z]+mu_gray[z]+mu_bright[z]
    return numerator/denominator
new_pic = np.zeros(im2.shape)
it = np.nditer(new_pic,flags = ['multi_index'],op_flags = ['readwrite'])
while not it.finished:
    z = im2[it.multi_index[0]][it.multi_index[1]]
    it[0] = defuzzification(z)
    it.iternext()
plt.imshow(new_pic)
plt.show()
```
