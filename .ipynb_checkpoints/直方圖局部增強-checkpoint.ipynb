```python
%matplotlib inline
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
from scipy.linalg import solve
import seaborn as sns
sns.set_style("white")
im = Image.open('fa.jpeg')
im = np.asarray(im.convert('L'))
plt.imshow(im)
plt.gray()
plt.show()
ax = sns.distplot(im.flatten(), color="y")
# neighber_area = im[0:10,0:2]
# it = np.nditer(neighber_area, flags=['multi_index'], op_flags=['readonly'])
# while not it.finished:
#     print(it.multi_index[0],it.multi_index[1])
#     it.iternext()
```

## 先計算整體的平均值、變異數。

```python
# 平均數
L = 256
M = np.average(im.flatten())
Sigma = np.sqrt(np.var(im.flatten()))
E = 4
k0 = 0.4
k1 = 0.02
k2 = 0.4
L = 256
print('M:',M,'Sigma:',Sigma,'E:',E)
```

```python
# 新圖片空矩陣
new_pic = np.copy(im)
new_pic.astype(np.uint8)
filter_size = (11,11)
half_size = np.uint((filter_size[0]-1)/2)

for i in range(half_size,new_pic.shape[0]-half_size):
    for j in range(half_size,new_pic.shape[1]-half_size):
        #計算鄰域的統計量
        neighber_area = new_pic[i-half_size:i+half_size+1,j-half_size:j+half_size+1]
        it = np.nditer(neighber_area, flags=['multi_index'], op_flags=['readwrite'])
        m = np.average(neighber_area.flatten())
        sigma = np.sqrt(np.var(neighber_area.flatten()))
        if m <= k0*M and sigma>=k1*Sigma and sigma<=k2*Sigma:
            hist, bin_edges = np.histogram(neighber_area.flatten(), density=False,bins=L,range=(0,L))
            Accumulate_hist = np.add.accumulate(hist)
            S = (L-1)/(neighber_area.shape[0]*neighber_area.shape[1])*Accumulate_hist
            # 更新區域內所有的強度
            while not it.finished:
                it[0] = S[it[0]]
                it.iternext()
        else:
            continue
plt.imshow(new_pic.astype(np.uint8))
plt.gray()
plt.show()
```
