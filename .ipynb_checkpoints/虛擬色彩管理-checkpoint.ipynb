```python
%matplotlib inline
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
from scipy.linalg import solve
import seaborn as sns
import pandas as pd
sns.set_style("white")
im = Image.open('C:/Users/user/Documents/DIP/DIP3E_Original_Images_CH06/Fig0620(a)(picker_phantom).tif')
im = np.asarray(im.convert('L'))
im2 = Image.open('C:/Users/user/Documents/DIP/DIP3E_Original_Images_CH06/Fig0627(a)(WashingtonDC Band3-RED).TIF')
im2 = np.asarray(im2.convert('L'))
plt.imshow(im)
plt.show()
plt.imshow(im2)
plt.show()
plt.figure()
ax = sns.distplot(im.flatten(), color="y")
plt.figure()
ax2 = sns.distplot(im2.flatten(), color="r")
```

```python
h = ['#FF0000','#FF3A00','#FF6800','#FFB900','#FFFB00','#C6FF00','#7DFF00','#3BFF00','#00FF29','#00FF7A','#00FFD3',
     '#00BCFF','#007AFF','#0031FF','#0100FF','#5600FF','#A700FF','#E900FF']
color = np.zeros((len(h),1,3))


for i in range(len(h)):
    h_temp = h[i].lstrip('#')
    color[i] = np.array(list(int(h_temp[i:i+2], 16) for i in (0, 2 ,4)))
```

```python
color_look_up = np.array(range(256))
bins = np.array(range(1,len(h)+1))/len(h)*256
inds = np.digitize(color_look_up,bins)
inds
```

```python

new_pic = np.zeros((im.shape[0],im.shape[1],3))
for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        new_pic[i][j] = color[inds[im[i][j]]]
 
  
plt.imshow(new_pic)

plt.show()
```

```python
def fr(t):
    return np.abs(np.sin(np.pi*t/100))*256
def fg(t):
    return np.abs(np.sin(np.pi*(t-10)/100))*(256)
def fb(t):
    return np.abs(np.sin(np.pi*(t-30)/100))*256
tr = np.arange(0,256,1)
tg = np.arange(0,256,1)
tb = np.arange(0,256,1)
plt.figure(1)
plt.subplot(311)
plt.plot(tr, fr(tr), 'r')
plt.subplot(312)
plt.plot(tg, fg(tg), 'g')
plt.subplot(313)
plt.plot(tb, fb(tb), 'b')
```

```python
im_r = im2.copy()
im_g = im2.copy()
im_b = im2.copy()
rfunc = np.vectorize(fr)
gfunc = np.vectorize(fg)
bfunc = np.vectorize(fb)
im_r = rfunc(im_r)
im_g = gfunc(im_g)
im_b = bfunc(im_b)
output_img = np.stack((im_r,im_g,im_b),axis=-1)
plt.figure()
plt.imshow(output_img.astype(np.uint8))
plt.show()
```
